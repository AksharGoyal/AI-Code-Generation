## **Intro to GPT3**

GPT-3, or **G**enerative **P**retrained **T**ransformer of the third generation, is a neural network machine learning model trained using internet data to generate any sort of text. It was created by OpenAI and requires a small amount of text as input to generate vast quantities of pertinent and sophisticated machine-generated text. It can be used to for chatbots, **G**enerative **Q**uestion-**A**nwering (GQA), summarization, and much more.

The deep learning neural network of GPT-3 contains more than 175 billion machine learning parameters. Before GPT-3, Microsoft's Turing Natural Language Generation (NLG) model, which had 10 billion parameters, was the largest trained language model. GPT-3 is the world's largest neural network as of early 2021. As a consequence, GPT-3 is superior to all previous models in producing text that appears plausible enough to have been written by a human.

GPT-3 processes incoming text to execute a variety of tasks involving natural language. It employs both natural language generation and processing to comprehend and generate text in natural human language. Human-comprehensible content generation has historically been difficult for machines that do not comprehend the complexities and nuances of language. GPT-3 has been used to generate significant quantities of articles, poetry, stories, news reports, and dialogue from a small amount of input text.

## **Risks of GPT-3 Model**
* Plagiarism without references: Language models such as GPT-3 are becoming more accurate, and it may become difficult to distinguish machine-generated content from that written by a human. This may raise issues with copyright and plagiarism.
* Achieving precision: Despite its proficiency in mimicking the format of human-generated text, GPT-3 struggles in many applications with empirical accuracy.
* Bias: Language models are susceptible to bias from machine learning.



## **Limitations of GPT-3 Model**

* GPT-3 does not learn continuously: It has been pre-trained, so it lacks an ongoing long-term memory that learns from every interaction.
* Having a constrained input capacity: The input capacity of transformer architectures, including GPT-3, is limited. The user cannot provide a large amount of text as input for the output, which can limit the functionality of certain applications. The prompt limit for GPT-3 is approximately 2,048 tokens
* Slow time to deduce: GPT-3 also has a slow inference time because the model requires a lengthy time to generate results.
*Absence of explicability: GPT-3 is susceptible to the same issues as other neural networks, namely their inability to explain and interpret why particular inputs result in particular outputs.


## **Summary of GPT3 Model**

* **Corpus**: GPT-3 has been pre-trained on a corpus of text comprised of five significant datasets, including Common Crawl, WebText2, Books1, Books2, and Wikipedia. This corpus contains nearly a trillion words, enabling GPT-3 to perform a remarkable number of NLP tasks in a zero-shot setting, or without example data.

* **Prompt engineering**: The prompt is the text input you provide the model to generate the desired output. Engineering is to design an effective prompt, it is necessary to provide the model with sufficient task context in the text input. GPT-3 examines the input and endeavours to respond with the text it deems most appropriate.
